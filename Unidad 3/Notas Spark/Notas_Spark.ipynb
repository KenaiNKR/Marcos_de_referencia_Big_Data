```python
import findspark
findspark.init()
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql import SparkSession

import pandas as pd
import numpy as np
```

```python
spark = SparkSession.builder \
    .appName("Notas") \
    .getOrCreate()
```

```python
df = pd.read_excel('/content/MAESTRO DE NOTAS Correguido.xls')
df
```

```python
df_eliminar = ['Estudiante', 'Nombre docente','Unnamed: 8','Unnamed: 12']
```

```python
df.drop(df_eliminar, axis=1, inplace=True)
```

```python
df
```

```python
df = spark.createDataFrame(df)
df.printSchema()
df.show()
```

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler, StandardScaler
from pyspark.sql.functions import col, when
```

```python
# Reemplazar valores NaN por 0 en las columnas seleccionadas
df_filled = df.fillna({'Num_matricula': 0, 'Asistencia': 0.0, 'Nota final': 0.0})

# Seleccionar características para clustering
feature_columns = ['Num_matricula', 'Asistencia', 'Nota final']
assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")

# Transformar las características en un solo vector
df_assembled = assembler.transform(df_filled)

# Escalar los datos
scaler = StandardScaler(inputCol="features", outputCol="scaledFeatures", withMean=True, withStd=True)
scaler_model = scaler.fit(df_assembled)
df_scaled = scaler_model.transform(df_assembled)

# Mostrar el resultado
df_scaled.select("features", "scaledFeatures").show(truncate=False)
```

```python
from pyspark.ml.clustering import KMeans
```

```python
# Definir el número de clusters
k = 4
kmeans = KMeans(k=k, seed=0, featuresCol="scaledFeatures", predictionCol="Cluster")

# Ajustar el modelo
model = kmeans.fit(df_scaled)

# Realizar predicciones
df_with_clusters = model.transform(df_scaled)

# Mostrar el resultado
df_with_clusters.select("Num_matricula", "Asistencia", "Nota final", "Cluster").show(truncate=False)
```

```python
import matplotlib.pyplot as plt

# Convertir el DataFrame de Spark a Pandas
df_pandas = df_with_clusters.select('Asistencia', 'Nota final', 'Cluster').toPandas()

# Crear la visualización
plt.figure(figsize=(10, 6))
plt.scatter(df_pandas['Asistencia'], df_pandas['Nota final'], c=df_pandas['Cluster'], cmap='viridis', alpha=0.5)
plt.title("Clustering de Asistencia y Nota Final", fontsize=16)
plt.xlabel("Asistencia", fontsize=14)
plt.ylabel("Nota Final", fontsize=14)
plt.colorbar(label='Cluster')
plt.tight_layout()
plt.show()
```
